```python
## Data Preprocessing Template

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:, 0].values
y = dataset.iloc[:, 1].values

#print(len(dataset))

# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 0)

#no scaling as library does it

%whos
```

```python
# Fitting LinearRegression to training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
X_train = X_train.reshape(-1, 1) # Features can't be a vector, it should be a feature matrix, therefore, need to scale it
regressor = regressor.fit(X_train, y_train)
```

```python
# Predicting test set results

# reshape feature vector to feature matrix
X_test = X_test.reshape(-1, 1)
y_pred = regressor.predict(X_test)
y_pred, y_test
```

```python
# plotting training set and regression model
%matplotlib inline
plt.scatter(X_train, y_train, color = "red")
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Salary Vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()

```

```python
# plotting TEST set and regression model
%matplotlib inline
plt.scatter(X_test, y_test, color = "red")
plt.plot(X_test, y_pred, color='blue')
plt.title('Salary Vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
